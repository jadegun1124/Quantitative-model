```python
import ccxt

# 1. Initialize Upbit (USE YOUR NEW KEYS)
upbit = ccxt.upbit({
    'apiKey': 'Your upbit access key',
    'secret': 'Your upbit secret key',
})

ticker = 'BTC/KRW'
investment_krw = 5500.0

try:
    print("[*] Fetching current market price...")
    # Get the lowest available selling price
    orderbook = upbit.fetch_order_book(ticker)
    current_price = orderbook['asks'][0][0]
    
    # Calculate the exact amount of BTC to buy with 5,000 KRW
    btc_amount = investment_krw / current_price
    
    print(f"[*] Placing Limit Buy Order for {btc_amount:.8f} BTC at {current_price:,.0f} KRW...")
    
    # 2. Execute the Order
    order = upbit.create_limit_buy_order(ticker, btc_amount, current_price)
    
    print("\n[+] SUCCESS! Order executed.")
    print(f"Order ID: {order['id']}")

except Exception as e:
    print("\n[-] FAILED. Check your API keys and permissions.")
    print(f"Error details: {e}")
```
